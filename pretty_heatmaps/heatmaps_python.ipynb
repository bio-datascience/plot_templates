```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Set plot style
sns.set_theme(style="white")
```

```python
count_data = pd.read_csv("./data/count.csv", index_col=0)
count_data
```

```python
col_annotation = pd.read_csv("./data/annotation.csv",index_col=0)
col_annotation
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(count_data)
plt.title("A basic heatmap of the count data")
plt.show()
```

```python
min_value = count_data[count_data > 0].min().min()
count_data_log = count_data.replace(0, min_value / 2)

# Log10 transform the data
count_data_log = np.log10(count_data_log)

# Create a diverging color palette
diverging_palette = sns.diverging_palette(240, 20, as_cmap=True)

plt.figure(figsize=(8, 6))
sns.heatmap(count_data_log, cmap=diverging_palette, xticklabels=False, yticklabels=False)
plt.title("Log10 Relative Abundance")
plt.show()
```

```python

plt.figure(figsize=(8, 6))
sns.heatmap(count_data_log, 
            cmap=diverging_palette, 
            robust=True,
            xticklabels=False, 
            yticklabels=False)
plt.title("Log10 Relative Abundance with Robust Color Scale")
plt.show()
```

```python
col_colors1 = {'Feces': '#2ca02c', 'Skin': '#d62728', 'Tongue': '#9467bd',
                 'Soil': "#8c564b", 'Freshwater': "#e377c2", 'Freshwater (creek)': "#7f7f7f",
                 'Ocean': "#bcbd22", 'Sediment (estuary)': "#17becf", 'Mock': "#D3D3D3"}
col_colors = col_annotation['SampleType'].map(col_colors1)
plt.figure(figsize=(8, 6))
# Create the clustermap
g = sns.clustermap(count_data_log, 
                   cmap=diverging_palette, 
                   robust=True,
                   col_colors=col_colors, 
                   col_cluster=True,
                   row_cluster=True,
                   figsize=(9, 8),
            xticklabels=False, 
            yticklabels=False)

# Add a title to the figure
g.figure.suptitle('Clustered Heatmap with Annotations', y=1.02)

# Add axis titles
g.ax_heatmap.set_xlabel("Samples")
g.ax_heatmap.set_ylabel("OTUs")

# Add a legend for the annotation
from matplotlib.patches import Patch
handles = [Patch(facecolor=col_colors1[name]) for name in col_colors1]
plt.legend(handles, col_colors1, title='SampleType', bbox_to_anchor=(1, 1), 
           bbox_transform=plt.gcf().transFigure, loc='upper right')

plt.show()
```
